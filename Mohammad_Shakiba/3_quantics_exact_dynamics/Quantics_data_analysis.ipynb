```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
pop_1 = np.loadtxt('model_2/pop_1.txt')
pop_2 = np.loadtxt('model_2/pop_2.txt')
plt.plot(pop_1, color='red', label='S$_1$')
plt.plot(pop_2, color='blue', label='S$_2$')
plt.ylabel('Population')
plt.xlabel('Time, fs')
plt.title('Holstein Model 2')
plt.legend()
plt.tight_layout()
```

```python
%matplotlib notebook
ene_1 = np.loadtxt('model_2/ene_1.txt')
ene_2 = np.loadtxt('model_2/ene_2.txt')
plt.plot(ene_1, color='red', label='E$_1$')
plt.plot(ene_2, color='blue', label='E$_2$')
plt.plot(ene_1+ene_2, color='green', label='Total energy')
plt.ylabel('Energy, eV')
plt.xlabel('Time, fs')
plt.title('Holstein Model 2 energies')
plt.legend()
plt.tight_layout()
```

```python
%matplotlib notebook
auto_corr = np.loadtxt('model_2/auto')
plt.plot(auto_corr[:,0], auto_corr[:,3])
plt.ylabel('|I($\omega$)|')
plt.xlabel('Time, fs')
plt.title('Holstein Model 2')
```

```python
%matplotlib notebook
q = np.loadtxt('model_2/q.txt')
plt.plot(q[0::2][1:], label='S$_1$', color='red')
plt.plot(q[1::2][1:], label='S$_2$', color='blue')
plt.ylabel('Position, a.u.')
plt.xlabel('Time, fs')
plt.title('Position vs time')
plt.legend()
plt.tight_layout()
```

```python
%matplotlib notebook
t = 200
q = np.loadtxt('model_2/q.txt')
plt.scatter(q[0::2][1:t], ene_1[1:t], label='S$_1$', color='red', s=1.5)
plt.scatter(q[1::2][1:t], ene_2[1:t], label='S$_2$', color='blue', s=1.5)
plt.ylabel('Energy, eV')
plt.xlabel('Position, a.u.')
plt.title('Position and energy')
plt.legend()
plt.tight_layout()
```
